```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('epa-sea-level.csv')
```

```python
res = stats.linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
x_pred = pd.Series(i for i in range(1880, 2050))
```

```python
x = df[df['Year'] >= 2000]['Year']
y = df[df['Year'] >= 2000]['CSIRO Adjusted Sea Level']
```

```python
res1 = stats.linregress(x, y)
x_pred1 = pd.Series(i for i in range(1880, 2050))
```

```python
plt.plot(df['Year'], df['CSIRO Adjusted Sea Level'], 'o', label='Data')
plt.plot(x_pred, res.intercept + res.slope*x_pred, 'r', label='fitted line')
plt.legend()
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.show()
```

```python
plt.plot(df['Year'], df['CSIRO Adjusted Sea Level'], 'o', label='Data')
plt.plot(x_pred1, res1.intercept + res1.slope*x_pred1, 'r', label='fitted line')
plt.legend()
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.show()
```
